```python
import cv2
from cvzone.HandTrackingModule import HandDetector
```

```python
from cvzone.ClassificationModule import Classifier
```

```python
import numpy as np
import math

```

```python
cap=cv2.VideoCapture(0)
detector=HandDetector(maxHands=2)
classifier=Classifier("Model/keras_model.h5","Model/labels.txt")
```

```python
offset=30
imgSize=640
```

```python
Labels=["Bird","Zero","Tiger","Snake","Rat","Ram","Ox","Monkey","Horse","Hare","Dragon","Dog","Boar"]
```

```python
while True:
    success, img = cap.read()
    imgOutput=img.copy()
    hands, img = detector.findHands(img)
    
    if hands and len(hands) >= 2:
        hand = hands[1]
        x, y, w, h = hand['bbox']
        

        imgWhite = np.ones((imgSize, imgSize, 3), np.uint8) * 255
        imgCrop = img[y - offset:y + h + offset, x - offset:x + w + offset]
        
        aspectRatio = h / w
        
        if aspectRatio > 1:
            k = imgSize / h
            wCalculated = math.ceil(k * w)
            imgReSize = cv2.resize(imgCrop, (wCalculated, imgSize))
            imgWhite[:, (imgSize - wCalculated) // 2:(imgSize - wCalculated) // 2 + wCalculated] = imgReSize
            prediction, index= classifier.getPrediction(imgWhite, draw=False)
            print(prediction,index)
            
            
        else:
            k = imgSize / w
            hCalculated = math.ceil(k * h)
            imgReSize = cv2.resize(imgCrop, (imgSize, hCalculated))
            imgWhite[(imgSize - hCalculated) // 2:(imgSize - hCalculated) // 2 + hCalculated, :] = imgReSize
            prediction, index= classifier.getPrediction(imgWhite, draw=False)
        
        cv2.rectangle(imgOutput,(x-offset,y-offset-50),(x-offset+150,y-offset-50+50),(255,0,255),cv2.FILLED)
        cv2.putText(imgOutput,Labels[index],(x,y-40),cv2.FONT_HERSHEY_COMPLEX,1,(255,255,255),2)
        cv2.rectangle(imgOutput,(x-offset,y-offset),(x+w+offset,y+h+offset),(255,0,255),4)
        
        cv2.imshow("ImageCrop", imgCrop)
        cv2.imshow("ImageWhite", imgWhite)

    cv2.imshow("Image", imgOutput) 
    cv2.waitKey(1)
```
